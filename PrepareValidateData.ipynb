```python
import pickle, gzip
import pandas as pd
```

```python
with gzip.open('input/stanfordData_test_p5.nlp', 'rb') as fout:
    store = 1999999
    while True:
        try:
            tmp = pickle.load(fout)
            if (tmp['id'] != store+1):
                print(tmp['id']-1)
            store = tmp['id']
        except EOFError:
            print("EOF")
            break

print("Done!")
```
